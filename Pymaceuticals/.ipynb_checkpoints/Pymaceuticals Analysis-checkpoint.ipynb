# Pymaceuticals Analysis

**Drug Effectivity**
* Among Capomulin, Infubinol, Ketapril, and Placebo treatments, a reduction of tumor volume was only observed for the Capomulin group.

**Tumor Volume Changes**
* Capomulin saw a 20% reduction in tumor volume within 45 days.
* Contrastingly, the other three groups of interest saw an increase of 45% to 60% in tumor volume.
* A longer clinical study (1 yr) would be beneficial in determining whether further reduction in volume will be observed for Capomulin trials.

**Metastatic Spread**
* Metastatic spread was seen to decrease by a factor of two in the Capomulin group compared to the Placebo group.
* The Infubinol group also saw a decrease in metastatic spread compared to the Placebo group. However, since tumor volume increased at a rate similar to the Placebo group, Infubinol may not be as good of a choice as Capomulin. 
* A future study with both Capomulin and Infubinol may be beneficial to observe interaction effects.

---

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_df = pd.read_csv(mouse_drug_data_to_load)
trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
full_df = pd.merge(trial_df, drug_df, on='Mouse ID', how='left')

# Display the data table for preview
full_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupby_obj = full_df.groupby(['Drug', 'Timepoint'])
mean_tumor_volume = groupby_obj['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
mean_tumor_volume_df = pd.DataFrame(mean_tumor_volume.reset_index())

# Preview DataFrame
mean_tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
volume_std_err = groupby_obj['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
volume_std_err_df = pd.DataFrame(volume_std_err.reset_index())

# Preview DataFrame
volume_std_err_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_volume_df2 = mean_tumor_volume_df.pivot(index='Timepoint', columns='Drug', values = 'Tumor Volume (mm3)')
volume_std_err_df2 = volume_std_err_df.pivot(index='Timepoint', columns='Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
mean_tumor_volume_df2.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_tumor_volume_df2.index, mean_tumor_volume_df2['Capomulin'], yerr=volume_std_err_df2['Capomulin'],\
            color = 'red', marker = 'o', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.errorbar(mean_tumor_volume_df2.index, mean_tumor_volume_df2['Infubinol'], yerr=volume_std_err_df2['Infubinol'],\
            color = 'blue', marker = '^', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.errorbar(mean_tumor_volume_df2.index, mean_tumor_volume_df2['Ketapril'], yerr=volume_std_err_df2['Ketapril'],\
            color = 'green', marker = 's', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.errorbar(mean_tumor_volume_df2.index, mean_tumor_volume_df2['Placebo'], yerr=volume_std_err_df2['Placebo'],\
            color = 'black', marker = 'd', markersize = 5, linewidth = 0.5, linestyle='dashed')

plt.title('Tumor Response to Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Time (Days)')
plt.legend(loc = 'best', fontsize = 8)
plt.grid()

# Save the Figure
plt.savefig('Tumor Response to Treatment.png', dpi = 200)

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metastatic_response = groupby_obj['Metastatic Sites'].mean()

# Convert to DataFrame
mean_metastatic_response_df = pd.DataFrame(mean_metastatic_response.reset_index())

# Preview DataFrame
pd.DataFrame(mean_metastatic_response).head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_std_err = groupby_obj['Metastatic Sites'].sem()

# Convert to DataFrame
metastatic_std_err_df = pd.DataFrame(metastatic_std_err.reset_index())

# Preview DataFrame
pd.DataFrame(metastatic_std_err).head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_metastatic_response_df2 = mean_metastatic_response_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
metastatic_std_err_df2 = metastatic_std_err_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')


# Preview that Reformatting worked
mean_metastatic_response_df2.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_metastatic_response_df2.index, mean_metastatic_response_df2['Capomulin'], yerr=metastatic_std_err_df2['Capomulin'],\
            color = 'red', marker = 'o', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.errorbar(mean_metastatic_response_df2.index, mean_metastatic_response_df2['Infubinol'], yerr=metastatic_std_err_df2['Infubinol'],\
            color = 'blue', marker = '^', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.errorbar(mean_metastatic_response_df2.index, mean_metastatic_response_df2['Ketapril'], yerr=metastatic_std_err_df2['Ketapril'],\
            color = 'green', marker = 's', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.errorbar(mean_metastatic_response_df2.index, mean_metastatic_response_df2['Placebo'], yerr=metastatic_std_err_df2['Placebo'],\
            color = 'black', marker = 'd', markersize = 5, linewidth = 0.5, linestyle='dashed')

plt.title('Metastatic Spread During Treatment')
plt.ylabel('Met. Sites')
plt.xlabel('Treatment Duration (Days)')
plt.legend(loc = 'upper left', fontsize = 8)
plt.grid()

# Save the Figure
plt.savefig('Metastatic Spread During Treatment.png', dpi = 200)

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint
survival = groupby_obj['Mouse ID'].count()

# Convert to DataFrame
survival_df = pd.DataFrame(survival.reset_index()).rename(columns = {'Mouse ID': 'Mouse Count'})

# Preview DataFrame
survival_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_df2 = survival_df.pivot(index='Timepoint', columns='Drug', values='Mouse Count')

# Preview the Data Frame
survival_df2.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(survival_df2.index, survival_df2['Capomulin']/survival_df2['Capomulin'][0]*100, color = 'red', marker = 'o', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.plot(survival_df2.index, survival_df2['Infubinol']/survival_df2['Infubinol'][0]*100, color = 'blue', marker = '^', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.plot(survival_df2.index, survival_df2['Ketapril']/survival_df2['Ketapril'][0]*100, color = 'green', marker = 's', markersize = 5, linewidth = 0.5, linestyle='dashed')
plt.plot(survival_df2.index, survival_df2['Placebo']/survival_df2['Placebo'][0]*100, color = 'black', marker = 'd', markersize = 5, linewidth = 0.5, linestyle='dashed')

plt.title('Survival During Treatment')
plt.ylabel('Survival Rate (%)')
plt.xlabel('Time (Days)')
plt.legend(loc = 'lower left', fontsize = 8)
plt.grid()

# Save the Figure
plt.savefig('Survival During Treatment.png', dpi = 200)

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
def to_percent(x):
    return x*100
last_row_index = len(mean_tumor_volume_df2) - 1
percent_changes_by_drug_df = mean_tumor_volume_df2.pct_change(periods=last_row_index).iloc[last_row_index,:].apply(to_percent)

# Display the data to confirm
percent_changes_by_drug_df
mean_tumor_volume_df2.pct_change(periods=last_row_index).iloc[last_row_index:last_row_index+1].apply(to_percent)
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_changes_tuple = percent_changes_by_drug_df['Capomulin'], percent_changes_by_drug_df['Infubinol'],\
percent_changes_by_drug_df['Ketapril'], percent_changes_by_drug_df['Placebo'] 
percent_changes_tuple

# Splice the data between passing and failing drugs
passing_changes = [x for x in percent_changes_tuple if x < 0]
failing_changes =[x for x in percent_changes_tuple if x > 0]

# Orient widths. Add labels, tick marks, etc.
fig = plt.figure()
plt.bar(-0.5, passing_changes, color='g', width=1, align = 'edge', alpha = 0.75)
plt.bar(np.arange(1,4), failing_changes, color='r', width=1, alpha = 0.75)
plt.xticks(np.arange(4), ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'])
plt.title('Tumor Change Over 45 Day Treatment', fontweight = 'bold')
plt.ylabel('% Tumor Volume Change', fontweight = 'bold')
plt.xlabel('Drug', fontweight = 'bold')
plt.xlim(-0.7,3.7)
plt.ylim(-30,70)
plt.grid()

# Label the percentages of changes
def format_percent(x):
    return str(int(round(x)))+'%'

# Bar value labels
plt.text(-0.2, -6, format_percent(percent_changes_tuple[0]), color = 'w', fontweight = 'bold')
for a,b in zip(np.arange(1,4), percent_changes_tuple[1:4]):
    plt.text(a-0.1, 3, format_percent(b), color = 'w', fontweight = 'bold')
    
# Save the Figure
plt.savefig('Tumor Change Over 45 Day Treatment.png', dpi = 200)

# Show the Figure
fig.show()
```
